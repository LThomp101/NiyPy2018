```python
# Transformed
B = NaturalCare.append(Protective)
C = B.append(Non)
FullSampleDataTransformed = C.merge(MasterT, on='Sample ID', how='inner')
FullSampleData= C.merge(Master, on='Sample ID', how='inner')
cols = ['Methylparaben','Ethylparaben','n-Propylparaben','iso-Propylparaben','n-Butylparaben','iso-Butylparaben','Benzylparaben','BPA']
SampleDataTransformed.to_csv('FullT.csv')
SampleDataTransformed.tail()
```

```python
#NON Black
def q75(x):
    return x.quantile(0.75)
def q25(x):
    return x.quantile(0.25)

sns.set_theme(style="white", palette="Paired")
sns.light_palette("seagreen", as_cmap=True)

#ax = sns.boxplot(x="Category", y="Totals", data=SampleDataTransformed)
#ax = sns.swarmplot(x="Category", y="Totals", data=SampleDataTransformed, color=".05")
options = ['Non','Protective','NaturalCare']
Non = FullSampleData[FullSampleData['Category'].isin(options)]
NonTransformed = FullSampleDataTransformed[FullSampleDataTransformed['Category'].isin(options)]

ax = sns.violinplot(x="Totals", y="Category", data=Non,
                    inner=None, scale="count")
sns.despine(offset=10, trim=True);
ax = sns.swarmplot(x="Totals", y="Category", data=Non, color="white", edgecolor="gray")
Category = Non.groupby(['Category'], as_index=False).agg({"Totals": ["min", "max", "median", "mean","count","std", q75, q25]})

Category.columns = Category.columns.droplevel(0)
print(Category)

```

```python
#Scalp vs. Strands
def q75(x):
    return x.quantile(0.75)
def q25(x):
    return x.quantile(0.25)

options = ['Scalp','Strands']
ScalpStrands = SampleData[SampleData['Details'].isin(options)]
ScalpStrandsTransformed = SampleDataTransformed[SampleDataTransformed['Details'].isin(options)]

#BOXPLOT


ScalpStrands = ScalpStrands.groupby(['Details'], as_index=False).agg({"Totals": ["min", "max", "median", "mean","count","std", q75, q25]})

ScalpStrands.columns = ScalpStrands.columns.droplevel(0)
print(ScalpStrands)


ScalpStrands.to_csv('ScalpStrands.csv')

sns.set_theme(style="white", palette="Paired")
sns.despine(offset=10, trim=True);
sns.light_palette("seagreen", as_cmap=True)

ax = sns.violinplot(x="Totals", y="Details",
                    data= ScalpStrandsTransformed,
                    scale="count", palette="Set3")
sns.despine(offset=10, trim=True);
ax = sns.swarmplot(x="Totals", y="Details", data=ScalpStrandsTransformed, color="black", edgecolor="gray")
```

```python
# TRENDY VS LEGACY PRODUCTS
def q75(x):
    return x.quantile(0.75)
def q25(x):
    return x.quantile(0.25)

options = ['Yes','No']
Kids = SampleData[SampleData['Explicitly Kids'].isin(options)]
KidsTransformed = SampleDataTransformed[SampleDataTransformed['Explicitly Kids'].isin(options)]

#BOXPLOT


Kids = Kids.groupby(['Explicitly Kids'], as_index=False).agg({"Totals": ["min", "max", "median", "mean","count","std", q75, q25]})

Kids.columns = Kids.columns.droplevel(0)
print(Kids)


Kids.to_csv('Kids.csv')

sns.set_theme(style="white", palette="Paired")
sns.despine(offset=10, trim=True);
sns.light_palette("seagreen", as_cmap=True)

ax = sns.violinplot(x="Totals", y="Explicitly Kids",
                    data= KidsTransformed,
                    scale="count", palette="Blues")
sns.despine(offset=10, trim=True);
ax = sns.swarmplot(x="Totals", y="Explicitly Kids", data=KidsTransformed, color="black", edgecolor="gray")
```
